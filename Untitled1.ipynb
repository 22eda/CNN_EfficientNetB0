```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, ModelCheckpoint, EarlyStopping
import os
import zipfile
import shutil
import time
from google.colab import files

# --- 1. GÜVENLİ ZİP AÇMA (BadZipFile hatasını önler) ---
zip_yolu = '/content/data.zip'
extract_path = '/content/data'

# Dosya gerçekten var mı ve yüklendi mi kontrolü
if not os.path.exists(zip_yolu) or os.path.getsize(zip_yolu) < 1000:
    print(" HATA: 'data.zip' bulunamadı veya henüz tam yüklenmedi! Lütfen yüklemenin bitmesini BEKLE.")
    exit()

# Eski klasörü temizle
if os.path.exists(extract_path):
    shutil.rmtree(extract_path)
os.makedirs(extract_path)

print(" Zip dosyası güvenli yöntemle açılıyor...")
os.system(f'unzip -q "{zip_yolu}" -d "{extract_path}"')

# --- 2. KLASÖR YOLLARI ---
icerik = os.listdir(extract_path)
if 'train' in icerik:
    train_dir = os.path.join(extract_path, 'train')
    test_dir = os.path.join(extract_path, 'test')
elif 'DATASET' in icerik:
    train_dir = os.path.join(extract_path, 'DATASET', 'train')
    test_dir = os.path.join(extract_path, 'DATASET', 'test')
else:
    train_dir = os.path.join(extract_path, 'train')
    test_dir = os.path.join(extract_path, 'test')

if not os.path.exists(train_dir):
    print(f" HATA: Eğitim klasörü bulunamadı. Lütfen zip içeriğini kontrol et.")
    exit()

# --- 3. VERİ YÜKLEME ---
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)
val_datagen = ImageDataGenerator(rescale=1./255)

print("\n--- Veriler Hazırlanıyor ---")
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    color_mode="rgb",
    batch_size=32,
    class_mode='categorical',
    shuffle=True
)
validation_generator = val_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    color_mode="rgb",
    batch_size=32,
    class_mode='categorical'
)

# --- 4. MODEL MİMARİSİ ---
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = True

for layer in base_model.layers[:50]:
    layer.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = BatchNormalization()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(7, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

# --- 5. EĞİTİM ---
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

callbacks = [
    ReduceLROnPlateau(monitor='val_accuracy', factor=0.5, patience=2, verbose=1),
    ModelCheckpoint('balanced_model_final.keras', save_best_only=True, monitor='val_accuracy'),
    EarlyStopping(monitor='val_accuracy', patience=6, restore_best_weights=True)
]

print("\n BÜYÜK EĞİTİM BAŞLIYOR...")
history = model.fit(
    train_generator,
    epochs=25,
    validation_data=validation_generator,
    callbacks=callbacks
)

# --- 6. MODELİ KAYDETME VE OTOMATİK İNDİRME ---
dosya_ismi = 'bilim_senligi_final.h5'
model.save(dosya_ismi)
print(f"\n Eğitim bitti! En iyi model '{dosya_ismi}' adıyla kaydedildi.")

# Dosyanın diske tam yazılması için 3 saniye bekleme süresi
time.sleep(3)

if os.path.exists(dosya_ismi):
    print(" İndirme işlemi başlatılıyor...")
    print(" Lütfen tarayıcının üst kısmında çıkabilecek 'Dosya indirme iznine' ONAY verin.")
    files.download(dosya_ismi)
else:
    print(" HATA: Model dosyası kaydedilemedi!")
    print("Mevcut dosyalar:", os.listdir())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

print("\n📈 DEĞERLENDİRME VE GÖRSELLEŞTİRME AŞAMASINA GEÇİLİYOR...")

# 1. Test verisi için KESİNLİKLE shuffle=False olmalı!
# Yoksa tahmin sırası ile gerçek etiket sırası karışır.
eval_generator = val_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    color_mode="rgb",
    batch_size=32,
    class_mode='categorical',
    shuffle=False # <--- EN KRİTİK NOKTA
)

# 2. Modelden Tahminleri Alma
print("Bütün test verisi üzerinden tahminler hesaplanıyor, lütfen bekleyin...")
Y_pred = model.predict(eval_generator)
y_pred = np.argmax(Y_pred, axis=1)
y_true = eval_generator.classes

# Sınıf isimlerini otomatik alma ('angry', 'happy' vs.)
class_names = list(eval_generator.class_indices.keys())

# 3. İstatistiksel Sınıflandırma Raporu (Classification Report)
print("\n📊 SINIFLANDIRMA RAPORU (Classification Report):")
print(classification_report(y_true, y_pred, target_names=class_names))

# 4. Karmaşıklık Matrisi (Confusion Matrix) Görselleştirme
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Tahmin Edilen Duygu (Predicted)")
plt.ylabel("Gerçek Duygu (Actual)")
plt.title("Karmaşıklık Matrisi (Confusion Matrix)")
plt.show()


# 5. Eğitim ve Doğrulama Eğrileri (Loss & Accuracy Curves)
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(14, 5))

# Kayıp (Loss) Grafiği
plt.subplot(1, 2, 1)
plt.plot(epochs_range, loss, label='Eğitim Kaybı (Train Loss)', marker='o')
plt.plot(epochs_range, val_loss, label='Doğrulama Kaybı (Val Loss)', marker='o')
plt.title('Eğitim ve Doğrulama Kaybı (Loss)')
plt.xlabel('Epoch (Tur)')
plt.ylabel('Kayıp Değeri')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Başarı (Accuracy) Grafiği
plt.subplot(1, 2, 2)
plt.plot(epochs_range, acc, label='Eğitim Doğruluğu (Train Acc)', marker='o')
plt.plot(epochs_range, val_acc, label='Doğrulama Doğruluğu (Val Acc)', marker='o')
plt.title('Eğitim ve Doğrulama Doğruluğu (Accuracy)')
plt.xlabel('Epoch (Tur)')
plt.ylabel('Doğruluk Oranı')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import zipfile
import shutil

# --- 1. VERİYİ TEKRAR HAZIRLAMA ---
zip_yolu = '/content/data.zip'
extract_path = '/content/data'

if os.path.exists(zip_yolu):
    if not os.path.exists(extract_path):
        print(" Zip dosyası açılıyor...")
        os.system(f'unzip -q "{zip_yolu}" -d "{extract_path}"')
else:
    print(" HATA: 'data.zip' bulunamadı! Lütfen sol menüye yükle.")
    exit()

# Test klasörünü bulalım
icerik = os.listdir(extract_path)
if 'train' in icerik:
    test_dir = os.path.join(extract_path, 'test')
elif 'DATASET' in icerik:
    test_dir = os.path.join(extract_path, 'DATASET', 'test')
else:
    test_dir = os.path.join(extract_path, 'test')

# --- 2. MODELİ YÜKLEME ---
model_path = 'bilim_senligi_final.h5'

if not os.path.exists(model_path):
    print(" HATA: 'bilim_senligi_final.h5' bulunamadı! Lütfen sol menüye yükle.")
    exit()

print("\n Kaydedilen zeki model yükleniyor...")
# Keras modelimizi dosyadan okuyoruz
model = tf.keras.models.load_model(model_path)
print(" Model başarıyla yüklendi!")

# --- 3. TEST VERİSİNİ HAZIRLAMA (SHUFFLE=FALSE KRİTİK!) ---
val_datagen = ImageDataGenerator(rescale=1./255)

eval_generator = val_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    color_mode="rgb",
    batch_size=32,
    class_mode='categorical',
    shuffle=False # Tahminlerin sırasının karışmaması için ŞART
)

# --- 4. TAHMİN VE GÖRSELLEŞTİRME ---
print("\n Test verileri üzerinden sınav yapılıyor, lütfen bekleyin...")
Y_pred = model.predict(eval_generator)
y_pred = np.argmax(Y_pred, axis=1)
y_true = eval_generator.classes

class_names = list(eval_generator.class_indices.keys())

print("\n SINIFLANDIRMA RAPORU (Classification Report):")
print(classification_report(y_true, y_pred, target_names=class_names))

# Karmaşıklık Matrisi (Confusion Matrix)
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Tahmin Edilen Duygu (Predicted)")
plt.ylabel("Gerçek Duygu (Actual)")
plt.title("Karmaşıklık Matrisi - Duygu Analizi Modeli")
plt.show()
```
